```python
# Bronze ingestion: read CSVs from the UC Volume and write bronze tables

spark.sql("CREATE SCHEMA IF NOT EXISTS workspace.bronze")

base = "/Volumes/workspace/bronze/landing"
tables = ["customers","merchants","accounts","transactions","fx_rates"]
for name in tables:
    df = (spark.read
          .option("header",True)
          .option("inferSchema",True) # let Spark detect column types automatically
          .csv(f"{base}/{name}.csv"))
    (df.write.mode("overwrite") # replace the table each run (full batch load)
     .option("overwriteSchema","true") # fully replace the schema too
     .saveAsTable(f"workspace.bronze.bronze_{name}"))
    print(f"loaded bronze_{name}:{df.count()} rows")
spark.sql("SHOW TABLES IN workspace.bronze").show()
```
