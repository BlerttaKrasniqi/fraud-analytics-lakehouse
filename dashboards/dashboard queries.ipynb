```sql

-- ============================================================
-- Fraud Analytics Dashboard - source queries
-- Each query below backs one dataset/tile in the Databricks
-- AI/BI dashboard. Reads from the gold layer (workspace.gold).
-- ============================================================
 
-- Dataset: kpis  ->  four Counter tiles
-- Headline executive numbers: total txns, flagged count, % flagged, $ at risk
SELECT *
FROM workspace.gold.gold_fraud_kpis;
 
 
-- Dataset: by_category  ->  Bar chart (X = category, Y = flagged_pct)
-- Flagged rate per merchant category, highest first
SELECT category,
       txn_count,
       flagged,
       flagged_pct
FROM workspace.gold.gold_risk_by_category
ORDER BY flagged_pct DESC;
 
 
-- Dataset: by_level  ->  Pie chart (label = risk_level, value = txn_count)
-- Distribution of transactions across Low / Medium / High risk
SELECT risk_level,
       txn_count
FROM workspace.gold.gold_risk_level_breakdown;
 
 
-- Dataset: trend  ->  Line chart (X = day, Y = flagged / txn_count)
-- Daily transaction volume vs flagged volume over time
SELECT day,
       txn_count,
       flagged
FROM workspace.gold.gold_daily_risk_trend
ORDER BY day;
 
 
-- Dataset: flagged  ->  Table tile
-- The risky transactions with human-readable reasons (feeds review + chatbot)
SELECT transaction_id,
       amount_usd,
       risk_level,
       risk_reason
FROM workspace.gold.gold_flagged_transactions
ORDER BY risk_score DESC;
```
